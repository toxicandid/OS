### Titanic Data preprocessing
https://www.kaggle.com/datasets/yasserh/titanic-dataset?resource=download

```python
import numpy as np
import pandas as pd
```

```python
dataset= pd.read_csv('Titanic-Dataset.csv')
```

```python
dataset.head()
```

```python
dataset.isnull().sum()
```

* ## nanmean :
  It is a function from the NumPy library that calculates the mean of an array, ignoring NaN values. It takes an input array and returns the mean of its non-NaN elements

```python
mean=int(np.nanmean(dataset['Age'],axis =0))
mean
```

Filling this mean value in the missing columns

```python
dataset['Age']=dataset['Age'].fillna(mean)
```

```python
# Letâ€™s check if there is any value missing in the Age column

dataset['Age'].isnull().sum()
```

## Calculate the survival rate for each passenger class

```python
survival_rate_by_class = []
for i in range(1, 4):
    passengers_in_class = dataset[dataset.iloc[:, 2] == i]
    survival_rate = np.mean(passengers_in_class.iloc[:, 1])
    survival_rate_by_class.append((f'Class {i}', survival_rate))
survival_rate_by_class
```

### For further illustration letâ€™s use another dataset called mtcars.csv. It consists of 11 features of 32 automobiles from the 1974 Motor Trend US magazine.

```python
dataframe=pd.read_csv('mtcars.csv')
dataframe.head()
```

##Standardization:

Standardization is a technique used to rescale data so that it has zero mean and unit variance.

We can standardize this data by separately calculating mean and standard deviation using following:

```python
mean=np.mean(dataframe.iloc[:,1:],axis=0)
mean
```

```python
std=np.std(dataframe.iloc[:,1:],axis=0)
std
```

```python
standardise_data=(dataframe.iloc[:,1:]- mean)/std
standardise_data.head()
```

## Scaling:

Scaling is a technique used to rescale data to a specific range.



```python
min=np.min(dataframe.iloc[:,1:],axis=0)
min
```

```python
max=np.max(dataframe.iloc[:,1:],axis=0)
max
```

Then, the scaled data is calculated by subtracting the minimum value from the data, dividing by the range of the data (max_val â€” min_val).

```python
scaled_data=(dataframe.iloc[:,1:]- min)/(max-min)
scaled_data.head() #This will give you a scaled version of your data between 0 and 1
```

## Normalization:

Normalization is a technique used to rescale data to have a unit norm.

* NumPyâ€™s linalg module provides several functions for computing norms, such as norm and linalg.norm. The np.linalg.norm function calculates the L2 norm of the data. Then, the normalized data is calculated by dividing the data by the L2 norm.

```python
norm=np.linalg.norm(dataframe.iloc[:,1:])
normalised_data=(dataframe.iloc[:,1:])/norm
```
